```python
#Alttaki iki komut aynı işlevi görüyor sadece yazımı farklı.
#data.Attack.max() 
data["Attack"].max()
```

```python
data[data["Attack"] == 190] #en yüksek atak gücün kime ait olduğu ve diğer özelliklerini gösterdik.
```

```python
data[data["Attack"] == 190]["Type 1"] #Atağı 190 olan karakterin sadece özelliğin ne olduğunu çağırdık.
```

```python
data[data["Attack"] == 190]["Type 1"].iloc[0]  #Atağı 190 olan karakterin sadece özelliğin ne olduğunu çağırdık. iloc ile sade gösterim yaptık.
```

```python
data[data["Name"] == "Pikachu"]["Defense"].iloc[0] #Pikachu'nun defans gücünün kaç olduğunu ekrana yazdırdık.
```

```python
data.groupby("Type 1") #type 1'i grupladık.
```

```python
data.groupby("Type 1").mean(numeric_only=True) #Type 1 sütunundakilere göre ortalamasını aldırdık. 
                                               #mean kısmı hata verdi çünkü string bir ifade olduğu için ortalamasını alamdı.
                                               #bu yüzden numeric olarak tanımladık.
```

```python
data.groupby("Type 1").mean(numeric_only=True).max() #type 1 sütunundaki ortalamaları aldırdıktan sonra totalde sütunların en güçlü olan sayıları ekrana yazdırdık.
```

```python
data["Type 1"].nunique() #Type 1 sütununda kaç farklı özellik olduğunu ekrana yazdırdık.
```
