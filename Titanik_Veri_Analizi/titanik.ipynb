```python
def degiskenler(variable):
    cat=train_data[variable]
    sayi=cat.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(sayi.index,sayi)
    plt.xticks(sayi.index,sayi.index.values)
    plt.ylabel("frekans")
    plt.title(variable)
    plt.show()
    print(variable,sayi)
```

```python
#Survived = Kurtulan(1) ve Kurtulamayan(0)
#Sex = Cinsiyet
#Pclass = Bilet kategorisi
#Embarked= hangi hangardan kaÃ§ kiÅŸi binmiÅŸ
#SibSp = yanÄ±nda bir eÅŸ ya da kardeÅŸi var mÄ±
#Parch = ailesinden biri var mÄ±
category=["Survived","Sex","Pclass","Embarked","SibSp","Parch"]
```

```python
for i in category:
    degiskenler(i)
```

```python
category2 = ["Cabin","Name","Ticket"]
for i in category2:
    print(train_data[i].value_counts())
```

```python
def hist(variable):
    plt.figure(figsize=(9,3))
    plt.hist(train_data[variable])
    plt.xlabel(variable)
    plt.ylabel("frekans")
    plt.title(variable)
    plt.show()
```

```python
sayisal=["Fare","Age","PassengerId"]
for i in sayisal:
    hist(i)
```
